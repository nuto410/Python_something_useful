```python
# image merge axis=0
import cv2
import numpy as np
import matplotlib.pyplot as plt

#read the image file
img1=cv2.imread(r'C:\Users\USER\Downloads\01.jpeg')
img2=cv2.imread(r'C:\Users\USER\Downloads\02.jpeg')
print('img1 size ', img1.shape)
print('img2 size ', img2.shape)
#resize the biger image
img1 = cv2.resize(img1, (480, 640), interpolation = cv2.INTER_AREA)

print('img1 size ', img1.shape)
print('img2 size ', img2.shape)
#merge the two image
image = np.concatenate((img1, img2),axis=0)

#resize the image
image = cv2.resize(image, (400,800), interpolation = cv2.INTER_AREA)
# #show the image on screen
# cv2.imshow('image',image)
# #keep the pic
# cv2.waitKey(0)

#save pic
cv2.imwrite(r'C:\Users\USER\Downloads\03.jpeg', image)
#show the merge image
img3 = cv2.imread(r'C:\Users\USER\Downloads\03.jpeg')
plt.imshow(img3)
#remove the ticks
plt.xticks([])
plt.yticks([])
```
