```python
#threading版本，前期很快，後期變慢。整體速度比asyncio慢
import re
import requests as rq
import os
from bs4 import BeautifulSoup
from tqdm import tqdm
import threading

#去除特定文字
def remove_specific_text_keep_spaces(input_string, text_to_remove):
    # 將 text_to_remove 分割成單個字符，並在每個字符之間插入 "\s*"，表示可以匹配任意多個空格
    pattern_str = " *".join(re.escape(char) for char in text_to_remove)
    pattern = re.compile(pattern_str)
    return pattern.sub("", input_string)

def get_text_data(a, num, text_data, lock):
    a_href = a.get('href')
    if a_href[:10]=='/Book/Read' and len(a_href)>11:
        response = rq.get('https://tw.hjwzw.com'+str(a_href)) # 用 requests 的 get 方法把網頁抓下來
        html_doc = response.text # text 屬性就是 html 檔案
        soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器
        text_to_remove="請記住本站域名: 黃金屋"
        search_text = remove_specific_text_keep_spaces(soup.text[750:-500],text_to_remove=text_to_remove)
        with lock:
            text_data[num] = search_text

def webcrawer_golderhouse(num, text_data, lock):
    url=f'https://tw.hjwzw.com/Book/Chapter/{num}'
    response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
    html_doc = response.text # text 屬性就是 html 檔案
    soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器
    all_a = soup.find_all("a")
    threads = []
    for a in all_a:
        thread = threading.Thread(target=get_text_data, args=(a, num, text_data, lock))
        thread.start()
        threads.append(thread)
    
    for thread in threads:
        thread.join()

def save_text_data(num, text_data):
    with open(f'{num}.txt','w',encoding='utf-8') as f:
        f.write(text_data[num])

if __name__ == "__main__":
    url=f'https://tw.hjwzw.com/Book/Chapter/32808'
    response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
    html_doc = response.text # text 屬性就是 html 檔案
    soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器
    all_a = soup.find_all("a")
    num_chapters = len(all_a)
    text_data = ['' for _ in range(num_chapters)]
    lock = threading.Lock()
    threads = []
    for num in tqdm(range(1, num_chapters + 1)):
        thread = threading.Thread(target=webcrawer_golderhouse, args=(num, text_data, lock))
        thread.start()
        threads.append(thread)
    
    for thread in threads:
        thread.join()

    threads = []
    for num in range(1, num_chapters + 1):
        thread = threading.Thread(target=save_text_data, args=(num, text_data))
        thread.start()
        threads.append(thread)
    
    for thread in threads:
        thread.join()
```

```python
#asyncio版本，超快
import re
import aiohttp
import asyncio
import os
from bs4 import BeautifulSoup
from tqdm import tqdm

# # 去除特定文字
# def remove_specific_text_keep_spaces(input_string, text_to_remove):
#     pattern_str = " *".join(re.escape(char) for char in text_to_remove)
#     pattern = re.compile(pattern_str)
#     return pattern.sub("", input_string)

async def get_chapter_data(session, index, chapter_link, queue, title):
    async with session.get('https://tw.hjwzw.com' + chapter_link) as response:
        html_doc = await response.text()
        soup = BeautifulSoup(html_doc, "html.parser")
        # text_to_remove = "請記住本站域名: 黃金屋"
        # search_text = remove_specific_text_keep_spaces(soup.text[750:-500], text_to_remove=text_to_remove)
        # print(search_text)
        await queue.put((index, title, soup.text[750:-500]))

async def main():
    url = f'https://tw.hjwzw.com/Book/Chapter/37323'
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            html = await response.text()
            soup = BeautifulSoup(html, 'html.parser')
            chapters = [(i, a['href'], a.text) for i, a in enumerate(soup.find_all("a")) if a.get('href') and a['href'][:10] == '/Book/Read' and len(a['href']) > 11]

    print(f"章節數量：{len(chapters)}")
    # print("章節列表：")
    # for index, _, title in chapters:
    #     print(f"章節{index + 1}：{title}")

    queue = asyncio.Queue()
    async with aiohttp.ClientSession() as session:
        tasks = [get_chapter_data(session, index, link, queue, title) for index, link, title in tqdm(chapters)]
        await asyncio.gather(*tasks)

    text_data = []
    while not queue.empty():
        text_data.append(await queue.get())

    # 根據章節順序合併內容
    text_data.sort(key=lambda x: x[0])
    merged_text = ''
    for _, title, text in text_data:
        # print("title",title)
        # merged_text += f"{title}\n{text}\n\n"
        merged_text += f"{text}\n\n"
        # merged_text += f"{title}\n{text}\n\n"

    with open('37323.txt', 'w', encoding='utf-8') as f:
        f.write(merged_text)

await main()
#32808蠱真人
#35126大潑猴
#1890西遊記
#40054獵魔烹飪手冊
#45781玩家請上車
#48391道詭異仙
#47935無限之劇本殺
#35482原始戰記
#45239荒誕推演游戲
#46974舊日馭龍
#43853黎明醫生/瘟疫醫生
#36187超級神基因
#1695佛本是道
#43236大奉打更人
#30224最後人類
#36320輪回樂園
#40264我能回檔不死
#37323我的恐怖猛鬼樓
```
