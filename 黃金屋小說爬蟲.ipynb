```python
#https://ithelp.ithome.com.tw/articles/10186119
import requests as rq
import os
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://tw.hjwzw.com/Book/Chapter/32808'
response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
html_doc = response.text # text 屬性就是 html 檔案
soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器

# 一些屬性或方法
print(1,soup.title) # 把 tag 抓出來
print("---")
print(2,soup.title.name) # 把 title 的 tag 名稱抓出來
print("---")
print(3,soup.title.string) # 把 title tag 的內容欻出來
print("---")
print(4,soup.title.parent.name) # title tag 的上一層 tag
print("---")
print(5,soup.a) # 把第一個 <a></a> 抓出來
print("---")
print(6,soup.find_all('a')) # 把所有的 <a></a> 抓出來
```

```python
# 成品
import re
#去除特定文字
def remove_specific_text_keep_spaces(input_string, text_to_remove):
    pattern = re.compile(re.escape(text_to_remove))
    return pattern.sub("", input_string)


def webcrawer_golderhouse(num):
    import requests as rq
    import os
    from bs4 import BeautifulSoup
    from urllib.request import urlopen
    # 輸入書本在黃金屋中的編號(網頁連結)
    #EX:蠱真人
    #https://tw.hjwzw.com/Book/34789
    #編號:34789
    url=f'https://tw.hjwzw.com/Book/Chapter/{num}'
    response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
    html_doc = response.text # text 屬性就是 html 檔案
    soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器

    # print(soup.find("a")) # 第一個 <a></a>
    # print("---")
    # print(soup.find_all("a")) # 全部 <a></a>
    all_a = soup.find_all("a")
    # print(all_a)

    text_data=''
    for a in all_a:
        a_href = a.get('href')
        print(a_href)
        if a_href[:10]=='/Book/Read' and len(a_href)>11:
    #         text_links.append('https://tw.hjwzw.com'+str(a_href))

            response = rq.get('https://tw.hjwzw.com'+str(a_href)) # 用 requests 的 get 方法把網頁抓下來
            html_doc = response.text # text 屬性就是 html 檔案
            soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器
            #標題
            # print(soup.title.string) # 把 title tag 的內容欻出來

            #內文content
#             all_div = soup.find('div', attrs={'style':'font-size: 20px; line-height: 30px; word-wrap: break-word; table-layout: fixed; word-break: break-all; width: 750px; margin: 0 auto; text-indent: 2em;'})
#             text_data+=all_div.text
            text_to_remove="請記住本站域名: 黃金屋"
            search_text = remove_specific_text_keep_spaces(soup.text[750:-500],text_to_remove=text_to_remove)
            text_data+=search_text
    with open(f'{num}.txt','w',encoding='utf-8') as f:
        for i in range(len(text_data)):
            f.write(text_data[i])
    print('done')

```
